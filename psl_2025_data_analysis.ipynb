# Number of Matches won by defending or chasing 

```python
import plotly.graph_objects as go

Psl["won_by"] = (
    Psl["won_by"]
    .astype(str)
    .str.strip()
    .str.lower()
)

Psl = Psl[Psl["won_by"].isin(["runs", "wickets"])]
Psl["won_by"] = Psl["won_by"].replace({
    "runs": "Defending",
    "wickets": "Chasing"
})
won_by = Psl["won_by"].value_counts()
fig = go.Figure(
    data=[
        go.Pie(
            labels=won_by.index,
            values=won_by.values
        )
    ]
)

fig.update_layout(
    title="Number of Matches Won by Defending or Chasing"
)

fig.update_traces(
    hoverinfo="label+percent",
    textinfo="label+value+percent",
    textfont_size=18,
    marker=dict(
        colors=["red", "lightgreen"],
        line=dict(color="black", width=2)
    )
)

fig.show()
```
